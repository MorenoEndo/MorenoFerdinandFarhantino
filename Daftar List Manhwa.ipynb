#**BUSINESS UNDERSTANDING**

##**Latar Belakang**
Industri manhwa, atau komik asal Korea Selatan, telah mengalami perkembangan pesat dalam beberapa tahun terakhir. Dengan meningkatnya popularitas platform digital seperti Webtoon, banyak manhwa yang mendapatkan basis penggemar yang besar di seluruh dunia.
Pemahaman mengenai faktor-faktor yang berkontribusi terhadap kesuksesan sebuah manhwa menjadi penting bagi penerbit, kreator, maupun platform distribusi.

##**Tujuan Bisnis**
Tujuan utama dari analisis dataset ini adalah untuk memahami faktor-faktor yang mempengaruhi popularitas dan kesuksesan sebuah manhwa. Analisis ini dapat membantu berbagai pemangku kepentingan, seperti:

1. Penerbit dan Platform Digital → Menentukan strategi pemasaran yang lebih baik dengan memahami tren yang disukai pembaca.

2. Kreator Manhwa → Mengidentifikasi elemen cerita atau gaya seni yang lebih disukai agar dapat menciptakan karya yang menarik perhatian pembaca.

3. Pembaca dan Penggemar → Menemukan rekomendasi manhwa berdasarkan preferensi umum dan peringkat.


##**Permasalahan Bisnis**
Berdasarkan tujuan bisnis di atas, beberapa permasalahan yang perlu dianalisis adalah:

1. Identifikasi Tren Genre yang Populer

- Genre apa yang paling banyak diminati oleh pembaca?

- Genre mana yang memiliki skor tertinggi?

- Apakah genre tertentu memiliki kecenderungan untuk mendapatkan lebih banyak pembaca?


2. Hubungan antara Popularitas dan Skor

- Apakah manhwa dengan skor tinggi selalu memiliki popularitas tinggi?

- Apakah ada outlier, yaitu manhwa dengan skor tinggi tetapi tidak populer, atau sebaliknya?


3. Pengaruh Status Publikasi terhadap Minat Pembaca

- Apakah manhwa yang masih publishing lebih populer dibandingkan yang sudah finished?

- Apakah pembaca lebih tertarik untuk membaca manhwa yang sudah tamat atau yang masih berlanjut?


4. Korelasi antara Panjang Cerita dan Kesuksesan

- Apakah manhwa dengan lebih banyak chapter cenderung lebih populer?

- Berapa rata-rata jumlah bab yang dimiliki oleh manhwa dengan skor tertinggi?


5. Analisis Faktor yang Mempengaruhi Jumlah Favorit

- Faktor apa yang paling mempengaruhi jumlah favorit? (Genre, skor, popularitas, atau jumlah bab?)

- Apakah ada pola tertentu dalam preferensi pembaca terhadap manhwa yang dijadikan favorit?

##**Kriteria Keberhasilan**

2. Mengidentifikasi hubungan antara skor, popularitas, dan jumlah pembaca → Menjelaskan apakah kualitas cerita berkorelasi dengan jumlah pembaca.
