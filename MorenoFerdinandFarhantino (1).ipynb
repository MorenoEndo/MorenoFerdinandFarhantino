```python
import pandas as pd
```

##DATASET KONSUMSI LISTRIK ASRAMA

```python
df = pd.read_csv("/dataset_konsumsi_listrik_asrama.csv")

df
```

##MELIHAT MISSING VALUES

```python
print((df.isna().sum() / len(df)) * 100)
```

##MELAKUKAN PENGECEKAN NILAI KOSONG PADA JUMLAH PENGHUNI

```python
df[df['Jumlah_Penghuni'].isnull()]
```

###MENGATASI MASALAH

```python
df['Jumlah_Penghuni'].dropna().describe()
```

####MEAN

```python
df['Jumlah_Penghuni'] = df['Jumlah_Penghuni'].fillna(df['Jumlah_Penghuni'].dropna().mean())
```

##MENGECEK DUPLIKAT

```python
df[df.duplicated()]
```

###MENGATASI DATA YANG DUPLIKAT

```python
df = df.drop_duplicates()
```

```python
df
```

##MENGECEK OUTLIERS

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
df
```

###MENGATASI DENGAN CARA IMPUTASI

```python
columns_to_impute = ["Jumlah_Penghuni"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

####PENGECEKKAN ULANG

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```
